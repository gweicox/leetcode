### 1. Two Sum (Easy)
https://leetcode.com/problems/two-sum

Given an array of integers **nums** and an integer **target**, return *indices* of the two numbers such that they add up to ***target***.

You may assume that each input would have ***exactly*** **one solution**, and you may not use the same element twice.

You can return the answer in any order.

##### Example 1:
Input: nums = [2,7,11,15], target = 9 <br>
Output: [0,1] <br>
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

##### Example 2:
Input: nums = [3,2,4], target = 6 <br>
Output: [1,2]

##### Example 3:
Input: nums = [3,3], target = 6 <br>
Output: [0,1]
 

##### Constraints:
- 2 <= nums.length <= 104 <br>
- -109 <= nums[i] <= 109 <br>
- -109 <= target <= 109 <br>
- **Only one valid answer exists.**

**Follow-up:** Can you come up with an algorithm that is less than O(n2) time complexity?

```python
def two_sum0(nums, target):
    for i, num1 in enumerate(nums):
        for j, num2 in enumerate(nums[i+1:], i+1):
            if num1 + num2 == target:
                return [i, j]

# 2022.06.21
# Runtime: 5577 ms, faster than 15.12% of Python3 online submissions for Two Sum.
# Memory Usage: 14.9 MB, less than 95.13% of Python3 online submissions for Two Sum.
```

```python
def two_sum(nums, target):
    seen = {}
    for i, num1 in enumerate(nums):
        num2 = target - num1
        if num2 in seen:
            return [seen[num2], i]
        seen[num1] = i

# 2022.06.22      
# Runtime: 63 ms, faster than 93.77% of Python3 online submissions for Two Sum.
# Memory Usage: 15.2 MB, less than 49.76% of Python3 online submissions for Two Sum.
```
